## 最小代码示例：创建 Tensor 与查看属性

这个代码块用于观察 `创建 Tensor 与查看属性`。运行后请重点看输出 shape、loss 或图像结果是否符合预期。

```python
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt

torch.manual_seed(42)
np.random.seed(42)

x = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
z = torch.zeros(2, 3)
r = torch.randn(3, 2)

print("x:", x)
print("shape:", x.shape)
print("dtype:", x.dtype)
print("device:", x.device)
print("zeros:", z.shape)
print("randn:", r.shape)
```

## 最小代码示例：reshape、view 与 flatten

这个代码块用于观察 `reshape、view 与 flatten`。运行后请重点看输出 shape、loss 或图像结果是否符合预期。

```python
a = torch.arange(12)
b = a.view(3, 4)
c = b.reshape(2, 6)
d = b.flatten()

print("a:", a.shape)
print("b:", b.shape)
print("c:", c.shape)
print("d:", d.shape)
```

## 完整实验：广播与矩阵乘法

这个代码块用于观察 `广播与矩阵乘法`。运行后请重点看输出 shape、loss 或图像结果是否符合预期。

```python
x = torch.arange(12).float().view(3, 4)
bias = torch.tensor([10.0, 20.0, 30.0, 40.0])
w = torch.randn(4, 2)

print("x + bias:")
print(x + bias)
print("matmul shape:", (x @ w).shape)
```

## 实验观察

从运行结果可以看到，Tensor 的 shape 会随着 `view`、`reshape`、矩阵乘法而变化。初学者调试时应先打印 shape，再判断模型层是否匹配。

从运行结果可以看到，代码输出不只是“能跑”，还可以帮助判断模型是否在按预期学习。建议把观察写进实验记录，而不是只保留最终准确率。

## 常见错误

- shape 不匹配：先打印输入、输出和标签 shape。
- dtype 不正确：分类标签通常是 `torch.long`，回归目标通常是 float。
- 忘记 `optimizer.zero_grad()`：梯度会累积。
- 学习率不合理：过大震荡，过小收敛慢。
- 评估阶段忘记 `model.eval()` 或 `torch.no_grad()`。

调试时可以先缩小数据集，确认模型能否在小样本上过拟合。
