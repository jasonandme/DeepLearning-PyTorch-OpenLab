## 最小代码示例：运行 MultiheadAttention

这个代码块用于观察 `运行 MultiheadAttention`。运行后请重点看输出 shape、loss 或图像结果是否符合预期。

```python
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt

torch.manual_seed(42)
np.random.seed(42)

tokens = torch.randn(2, 5, 8)
mha = nn.MultiheadAttention(embed_dim=8, num_heads=2, batch_first=True)
out, weights = mha(tokens, tokens, tokens, need_weights=True)

print("tokens:", tokens.shape)
print("out:", out.shape)
print("weights:", weights.shape)
```

## 最小代码示例：使用 key_padding_mask

这个代码块用于观察 `使用 key_padding_mask`。运行后请重点看输出 shape、loss 或图像结果是否符合预期。

```python
key_padding_mask = torch.tensor([[False, False, False, True, True],
                                 [False, False, False, False, True]])
out_masked, weights_masked = mha(tokens, tokens, tokens, key_padding_mask=key_padding_mask, need_weights=True)
print("masked out:", out_masked.shape)
print("masked weights:", weights_masked.shape)
```

## 完整实验：权重可视化

这个代码块用于观察 `权重可视化`。运行后请重点看输出 shape、loss 或图像结果是否符合预期。

```python
plt.imshow(weights_masked[0].detach(), cmap="viridis")
plt.colorbar()
plt.title("Masked attention weights")
plt.xlabel("key position")
plt.ylabel("query position")
plt.show()
```

## 实验观察

`embed_dim` 必须能被 `num_heads` 整除。mask 可以让模型忽略 padding 或未来位置。

从运行结果可以看到，代码输出不只是“能跑”，还可以帮助判断模型是否在按预期学习。建议把观察写进实验记录，而不是只保留最终准确率。

## 常见错误

- shape 不匹配：先打印输入、输出和标签 shape。
- dtype 不正确：分类标签通常是 `torch.long`，回归目标通常是 float。
- 忘记 `optimizer.zero_grad()`：梯度会累积。
- 学习率不合理：过大震荡，过小收敛慢。
- 评估阶段忘记 `model.eval()` 或 `torch.no_grad()`。

调试时可以先缩小数据集，确认模型能否在小样本上过拟合。
