## 最小代码示例：构造小图像

这个代码块用于观察 `构造小图像`。运行后请重点看输出 shape、loss 或图像结果是否符合预期。

```python
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt

torch.manual_seed(42)
np.random.seed(42)

image = torch.zeros(1, 1, 8, 8)
image[:, :, 2:6, 3:5] = 1.0

plt.imshow(image[0, 0], cmap="gray")
plt.title("Toy image")
plt.axis("off")
plt.show()
print("image shape:", image.shape)
```

## 最小代码示例：卷积与池化

这个代码块用于观察 `卷积与池化`。运行后请重点看输出 shape、loss 或图像结果是否符合预期。

```python
conv = nn.Conv2d(1, 4, kernel_size=3, padding=1)
pool = nn.MaxPool2d(2)

with torch.no_grad():
    conv_out = torch.relu(conv(image))
    pool_out = pool(conv_out)

print("conv_out:", conv_out.shape)
print("pool_out:", pool_out.shape)
```

## 完整实验：可视化特征图

这个代码块用于观察 `可视化特征图`。运行后请重点看输出 shape、loss 或图像结果是否符合预期。

```python
plt.figure(figsize=(6, 3))
for i in range(4):
    plt.subplot(1, 4, i + 1)
    plt.imshow(conv_out[0, i], cmap="viridis")
    plt.title(f"map {i}")
    plt.axis("off")
plt.tight_layout()
plt.show()
```

## 实验观察

卷积层会产生多个特征图，池化会降低空间尺寸。CNN 调试的核心是持续跟踪 `[N, C, H, W]`。

从运行结果可以看到，代码输出不只是“能跑”，还可以帮助判断模型是否在按预期学习。建议把观察写进实验记录，而不是只保留最终准确率。

## 常见错误

- shape 不匹配：先打印输入、输出和标签 shape。
- dtype 不正确：分类标签通常是 `torch.long`，回归目标通常是 float。
- 忘记 `optimizer.zero_grad()`：梯度会累积。
- 学习率不合理：过大震荡，过小收敛慢。
- 评估阶段忘记 `model.eval()` 或 `torch.no_grad()`。

调试时可以先缩小数据集，确认模型能否在小样本上过拟合。
