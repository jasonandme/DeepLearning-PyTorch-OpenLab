## 最小代码示例：构造序列输入

这个代码块用于观察 `构造序列输入`。运行后请重点看输出 shape、loss 或图像结果是否符合预期。

```python
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt

torch.manual_seed(42)
np.random.seed(42)

batch = 4
seq_len = 6
features = 3
x = torch.randn(batch, seq_len, features)
print("x shape:", x.shape)
```

## 最小代码示例：运行 RNN

这个代码块用于观察 `运行 RNN`。运行后请重点看输出 shape、loss 或图像结果是否符合预期。

```python
rnn = nn.RNN(input_size=3, hidden_size=5, batch_first=True)
out, h_n = rnn(x)
print("out shape:", out.shape)
print("h_n shape:", h_n.shape)
print("last output shape:", out[:, -1, :].shape)
```

## 完整实验：用 RNN 做简单回归

这个代码块用于观察 `用 RNN 做简单回归`。运行后请重点看输出 shape、loss 或图像结果是否符合预期。

```python
series = torch.sin(torch.linspace(0, 12, 160)).view(-1, 1)
window = 8
xs, ys = [], []
for i in range(len(series) - window):
    xs.append(series[i:i+window])
    ys.append(series[i+window])
x_seq = torch.stack(xs)
y_seq = torch.stack(ys)

model = nn.Sequential(nn.RNN(1, 12, batch_first=True),)
print("sequence data:", x_seq.shape, y_seq.shape)
```

## 实验观察

RNN 输出包含每个时间步的输出和最终隐状态。很多任务会取最后时间步表示作为整个序列的摘要。

从运行结果可以看到，代码输出不只是“能跑”，还可以帮助判断模型是否在按预期学习。建议把观察写进实验记录，而不是只保留最终准确率。

## 常见错误

- shape 不匹配：先打印输入、输出和标签 shape。
- dtype 不正确：分类标签通常是 `torch.long`，回归目标通常是 float。
- 忘记 `optimizer.zero_grad()`：梯度会累积。
- 学习率不合理：过大震荡，过小收敛慢。
- 评估阶段忘记 `model.eval()` 或 `torch.no_grad()`。

调试时可以先缩小数据集，确认模型能否在小样本上过拟合。
