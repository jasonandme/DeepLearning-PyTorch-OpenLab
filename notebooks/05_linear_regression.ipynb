## 最小代码示例：生成数据

这个代码块用于观察 `生成数据`。运行后请重点看输出 shape、loss 或图像结果是否符合预期。

```python
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt

torch.manual_seed(42)
np.random.seed(42)

x = torch.linspace(-3, 3, 160).view(-1, 1)
y = 3.0 * x + 2.0 + 0.5 * torch.randn_like(x)

plt.figure(figsize=(6, 4))
plt.scatter(x.numpy(), y.numpy(), s=16, alpha=0.7)
plt.title("Linear regression data")
plt.grid(alpha=0.3)
plt.show()
```

## 最小代码示例：训练模型

这个代码块用于观察 `训练模型`。运行后请重点看输出 shape、loss 或图像结果是否符合预期。

```python
model = nn.Linear(1, 1)
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.05)
losses = []

for epoch in range(120):
    pred = model(x)
    loss = criterion(pred, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    losses.append(loss.item())

print("weight:", model.weight.item())
print("bias:", model.bias.item())
```

## 完整实验：观察结果

这个代码块用于观察 `观察结果`。运行后请重点看输出 shape、loss 或图像结果是否符合预期。

```python
with torch.no_grad():
    fitted = model(x)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(losses)
plt.title("Loss")
plt.grid(alpha=0.3)

plt.subplot(1, 2, 2)
plt.scatter(x.numpy(), y.numpy(), s=16, alpha=0.5)
plt.plot(x.numpy(), fitted.numpy(), color="red")
plt.title("Fitted line")
plt.grid(alpha=0.3)
plt.show()
```

## 实验观察

拟合线会贴近样本整体趋势，但不会穿过所有点。噪声存在时，合理模型学习的是主要关系而不是每个样本的扰动。

从运行结果可以看到，代码输出不只是“能跑”，还可以帮助判断模型是否在按预期学习。建议把观察写进实验记录，而不是只保留最终准确率。

## 常见错误

- shape 不匹配：先打印输入、输出和标签 shape。
- dtype 不正确：分类标签通常是 `torch.long`，回归目标通常是 float。
- 忘记 `optimizer.zero_grad()`：梯度会累积。
- 学习率不合理：过大震荡，过小收敛慢。
- 评估阶段忘记 `model.eval()` 或 `torch.no_grad()`。

调试时可以先缩小数据集，确认模型能否在小样本上过拟合。
